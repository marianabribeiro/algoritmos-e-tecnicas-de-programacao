### Resumo

O trabalho enquadra-se nos trabalhos da aula prática e o objetivo foi criar um gestor de frações que permitisse criar frações, simplificá-las, realizar operações matemáticas. Outro objetivo foi criar listas e fazer também operações de somas dos elementos de uma mesma lista ou de duas listas diferentes. Por fim, o programa consegue também encontrar o maior elemento de uma lista de frações.

### Código 

```python
def menu():
    print (""" O que pretende fazer?
           (1)Criar fração
           (2)Simplificar fração
           (3)Somar frações
           (4)Subtrair frações           
           (5)Multiplicar frações
           (6)Dividir frações
           (7)Gerar lista de n frações
           (8)Somar os elementos de uma lista de frações
           (9)Somar os elementos de duas listas de frações
           (10)Maior elemento de uma lista de frações
           (0)Sair""")
           
          
def criarFracao(num, denum):
    return (num, denum)


def verFracao(f):
    print(str(f[0]) + "/" + str(f[1]))


def mdc(a, b):
    if (int(a))<(int(b)):
        return mdc(b,a)
    elif (int(a))%(int(b))==0:
        return b
    else:
        return mdc((int(a)), (int(a))%(int(b)))
    
def simplificarFracao(f):
    num, denum = f
    a = mdc(num, denum)
    return ((int(num))/(int(a)), (int(denum))/(int(a)))

def somaFracoes(f1, f2):
    num1, denum1 = f1
    num2, denum2 = f2
    NUM = (int(num1))*(int(denum2)) + (int(num2))*(int(denum1))
    DENUM = (int(denum1))*(int(denum2))
    soma = simplificarFracao(criarFracao(NUM, DENUM))
    return soma

def subtrairFracoes(f1, f2):
    num1, denum1 = f1
    num2, denum2 = f2
    NUM = ((int(num1))*(int(denum2))) - ((int(num2))*(int(denum1)))
    DENUM = (int(denum1))*(int(denum2))
    subtrair = simplificarFracao(criarFracao(NUM, DENUM))
    return subtrair

    
def multiplicarFracoes(f1, f2):
    num1, denum1 = f1
    num2, denum2 = f2
    NUM = (int(num1))*(int(num2))
    DENUM = (int(denum1))*(int(denum2))
    multip = simplificarFracao(criarFracao(NUM, DENUM))
    return multip    
    
    
def dividirFracoes(f1, f2):
    num1, denum1 = f1
    num2, denum2 = f2
    NUM = (int(num1))*(int(denum2))
    DENUM = (int(denum1))*(int(num2))
    div = simplificarFracao(criarFracao(NUM, DENUM))
    return div    
    
    
def gerarLista():
    Li = []
    n = input("Quantas frações quer que a lista tenha?  ->")
    for i in range (int(n)):
        num = input("Introduza o numerador da fraçãao número " + (str(i+1)) +":")
        denum = input("Introduza o denominador da fração número " + (str(i+1)) +":")
        elemento=criarFracao(num, denum)
        Li.append(elemento)
    return Li


def somarElemLista(L):
    soma = (0,1)
    for elem in L:
        soma=somaFracoes(soma, elem)
    return soma
        

def somarListas(L1, L2):
    if (len(L1) != len(L2)):
        print("As listas tem tamanhos diferentes, não é possível somá-las")
    else:
        Lis = []
        for i in range(len(L1)):
            soma = somaFracoes(L1[i], L2[i])  
            Lis.append(soma)
    return Lis


def maiorElemLista(L):
    maxi=(0,1)
    for i in range (len(L)):
        nummaxi, denummaxi = maxi
        num, denum = L[i]
        nummaxitemp = (int(nummaxi))*(int(denum))
        numtemp = (int(num))*(int(denummaxi))
        if (int(numtemp))>(int(nummaxitemp)):
            maxi = L[i]
            
    return maxi
       
op="1"
while op!="0":
    menu()
    op = input("Introduza a opção:")
    if op == "1":
        n = input("Introduza o numerador:")
        d = input("Introduza o denominador: ")
        fracao = verFracao(criarFracao(n, d))
        
        
    elif op=="2":
        n = input("Introduza o numerador:")
        d = input("Introduza o denominador: ")
        
        fracaosimp =  simplificarFracao(criarFracao(n, d))
        print(verFracao(fracaosimp))
        
    elif op=="3":
        n1 = input("Introduza o numerador:")
        d1 = input("Introduza o denominador: ")
        fracao1 = criarFracao(n1, d1)
        
        n2 = input("Introduza o numerador:")
        d2 = input("Introduza o denominador: ")
        fracao2 = criarFracao(n2, d2)  
    
        soma =  somaFracoes(fracao1, fracao2)
        print("A soma é: ", soma)
        

    elif op=="4":
        n1 = input("Introduza o numerador:")
        d1 = input("Introduza o denominador: ")
        fracao1 = criarFracao(n1, d1)
        
        n2 = input("Introduza o numerador:")
        d2 = input("Introduza o denominador: ")
        fracao2 = criarFracao(n2, d2)  
        
        subtracao =subtrairFracoes(fracao1, fracao2)
        print("O resultado é: " , subtracao)
        
    elif op=="5":
        n1 = input("Introduza o numerador:")
        d1 = input("Introduza o denominador: ")
        fracao1 = criarFracao(n1, d1)
        
        n2 = input("Introduza o numerador:")
        d2 = input("Introduza o denominador: ")
        fracao2 = criarFracao(n2, d2)  
        
        multiplicacao =  multiplicarFracoes(fracao1, fracao2)
        print("o resultado é:", multiplicacao)


    elif op=="6":
        n1 = input("Introduza o numerador:")
        d1 = input("Introduza o denominador: ")
        fracao1 = criarFracao(n1, d1)
        
        n2 = input("Introduza o numerador:")
        d2 = input("Introduza o denominador: ")
        fracao2 = criarFracao(n2, d2)  
        
        
        divisao = ( dividirFracoes(fracao1, fracao2))
        print("O resultado é: ", divisao)
        

    elif op=="7":
        Lista = gerarLista()
        print(Lista)
        
    elif op=="8":
        Lista = gerarLista()
        print("A lista com a soma dos elementos das listas é: ", somarElemLista(Lista))   
    
    elif op=="9":
        Lista1 = gerarLista()
        Lista2 = gerarLista()
        print("A lista com a soma dos elementos das listas dadas é:", somarListas(Lista1, Lista2))
       
    elif op=="10":
        Lista = gerarLista()
        print("O maior elemento da lista é: ", maiorElemLista(Lista))
```
