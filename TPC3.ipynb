### Resumo

A realização do programa para ordenar listas através do algoritmo Bubble Sort enquadra-se no exercício de gestão de listas de números inteiros realizado na aula prática, também aprensentado neste ficheiro.

O programa Bubble Sort ordena a lista dada comparando os valores da lista, índice a índice, trocando-os de posição quando o valor do índice menor da lista é maior que o valor do índice maior.


### Código

```python
import random
from random import randint

def menu():
    print (""" O que pretende fazer?
           (1)Introduzir a lista à mão
           (2)Gerar uma lista com valores aleatórios de 0 a 100
           (3)Ordenar a lista - BubbleSort
           (4)Encontrar o maior elemento da lista
           (0)Sair""")


def criarListaManual():
    n=int(input("Introduza o numero de elementos:  "))
    L=[]
    for i in range (n):
        n=input("Introduza o elemento " + str(i+1) + ":")
        L.append(n)
    return L


def criarListaAleatoria():
    n=int(input("Introduza o numero de elementos:  "))
    L=[]
    for i in range (n):
        n=random.randint(0,100)
        L.append(n)
    return L


def bubbleSort(l):
    trocas=1
    while trocas!=0:
        i=1
        trocas=0
        while i < (len(l)):        
            if l[i-1]>l[i]:        
                temp= l[i-1]       
                l[i-1]=l[i]           
                l[i]=temp            
                trocas=+1
            i=i+1
    return l

def maiorElem(L):
    maxi=0
    for i in range (len(L)):
        if int(L[i])>int(maxi):
            maxi=L[i]
    
    return maxi
      
        
op="1"
while op!="0":
    menu()
    op=input("Introduza a opção:")
    if op=="1":
        lista=criarListaManual()
        print (lista)
        
    elif op=="2":
        lista=criarListaAleatoria()
        print (lista)
        
    elif op=="3":
        op2=input("""Deseja:
           (1)Introduzir a lista à mão
           (2)Gerar uma lista com valores aleatórios de 0 a 100
           Introduza a opção:""")
        if op2=="1":
            lista=criarListaManual()
            print ("Esta é a lista que vau ser ordenada: ", lista)
            listaord=bubbleSort(lista)
            print("A lista ordenada: ", listaord)
            
        elif op2=="2":
            lista=criarListaAleatoria()
            print ("Esta é a lista que vau ser ordenada: ", lista)
            listaord=bubbleSort(lista)
            print ("A lista ordenada: ", listaord)    
            
    elif op=="4":
        op2=input("""Deseja:
           (1)Introduzir a lista à mão
           (2)Gerar uma lista com valores aleatórios de 0 a 100
           Introduza a opção:""")
        if op2=="1":
            lista=criarListaManual()
            print ("Esta é a lista na qual se vai procurar o maior elemento: ", lista)
            elem=maiorElem(lista)
            print ("o maior elemento é: ", elem)
            
        elif op2=="2":
            lista=criarListaAleatoria()
            print ("Esta é a lista na qual se vai procurar o maior elemento: ", lista)
            elem=maiorElem(lista)
            print ("o maior elemento é: ", elem)
              
    
    
```
