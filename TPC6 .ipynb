```python
import matplotlib.pyplot as plt


def menu():
    print (""" O que pretende fazer?
           (1)Criar Polinómio
           (2)Calcular Polinómio
           (3)Calcular tabela
           (4)Simplificar polinómio
           (5)Derivar o polinómio
           (6)Desenhar gráfico
           (0)Sair""")

def criarTermo(coef, exp):
    return (coef, exp)

def criarPolinomio():
    return []

def inserirTermo(p, t):
    return p.append(t)

def verTermo(t):
    c, e = t
    if e == "0":
        return str(c)
    elif e == "1":
        return str(c) + "x"
    elif c == "1":
        return "x^" + str(e)
    else:
        return (str(c) + "x^" + str(e))
    
def verPolinomio(p):
    res = ""
    for t in p:
        if res == "":
            res = verTermo(t)
        else:   
            res = res + " + " + verTermo(t)
    print(res)
    
    
def calcPol(p, x):
    res = 0
    for (c,e) in p:
        res = res + (int(c)) * (int(x)) **(int( e))
    return res

def tabela(p, nlinhas):
    for i in range((int(nlinhas))):
        print(str(i) + " :: " + str(calcPol(p, (int(i)))))
        
def simplificarPol(p):
    ordenarPol(p)
    a,b=p[0]
    expmax=b
    L=[]
    poln=[]
    for i in range ((int(expmax))+1):
        for j in range(len(p)): 
            c,e=p[j]
            if (int(e))==(int(i)):
                L.append(c)
        soma=0
        for el in L:
            soma=soma+(int(el))
        if soma!=0:
            poln.append(criarTermo(soma, i))
        L=[]    
    ordenarPol(poln)
    verPolinomio(poln)  
    return poln
      
        
def chaveOrd(t):
    _,e = t  #ou c, e = t
    return e

def ordenarPol(p):
    p.sort(reverse=True, key=chaveOrd)

    
def derivarPol(p):
    res = []
    for (c,e) in p:
        if (int(e)) > 0:
            res.append(((int(c))*(int(e)), (int(e))-1))
    return res

def desenharGrafico(p, n):
    x=[]
    y=[]
    for i in range((int(n))):
        x.append(i)
        y.append(calcPol(p, (int(i))))
    
    plt.plot(x, y)
    plt.xlabel('Abcissas')
    plt.ylabel('Ordenadas')
    plt.title('O polinomio')
    plt.show()
    return x

      
op="1"
p=[]
while op!="0":
    menu()
    op=input("Introduza a opção:")
    
    if op=="1":
        n=input("Introduza o numero de termos do polinomio:")
        p=criarPolinomio()
        for i in range (int(n)):
            c=input("Introduza o coeficiente:")
            e=input("Introduza o expoente:")
            t=criarTermo(c, e)
            inserirTermo(p, t)
            verPolinomio(p)
        ordenarPol(p)
        print("=")
        verPolinomio(p)
        
    elif op=="2":
        if p==[]:
            n=input("Introduza o numero de termos do polinomio:")
            p=criarPolinomio()
            for i in range (int(n)):
                c=input("Introduza o coeficiente:")
                e=input("Introduza o expoente:")
                t=criarTermo(c, e)
                inserirTermo(p, t)
                verPolinomio(p)
            ordenarPol(p)
            print("O polinómio que vai ser usado é: ")
            verPolinomio(p)
            x=input("Introduza o valor de x:  ")
            res=calcPol(p, x)
            print("o valor do polinómio para o valor de x é: " , res)
        else:    
            print("O polinómio anterior é: ")
            verPolinomio(p)
            n=input("Pretende usar o polinomio anterior? responda s para sim, n para nao:  ")
            if n=="s":
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                x=input("Introduza o valor de x:  ")
                res=calcPol(p, x)
                print("o valor do polinómio para o valor de x é: " , res)
            elif n=="n":
                n=input("Introduza o numero de termos do polinomio:")
                p=criarPolinomio()
                for i in range (int(n)):
                    c=input("Introduza o coeficiente:")
                    e=input("Introduza o expoente:")
                    t=criarTermo(c, e)
                    inserirTermo(p, t)
                    verPolinomio(p)
                ordenarPol(p)
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                x=input("Introduza o valor de x:  ")
                res=calcPol(p, x)
                print("o valor do polinómio para o valor de x é: " , res)

        
    elif op=="3":
        if p==[]:
            n2=input("Introduza o numero de termos do polinomio:")
            p=criarPolinomio()
            for i in range (int(n2)):
                c=input("Introduza o coeficiente:")
                e=input("Introduza o expoente:")
                t=criarTermo(c, e)
                inserirTermo(p, t)
                verPolinomio(p)
            ordenarPol(p)
            print("O polinómio que vai ser usado é: ")
            verPolinomio(p)
            x=input("Introduza o numero de linhas a calcular:  ")
            tabela(p, x)
        
        else:
            
            print("O polinómio anterior é: ")
            verPolinomio(p)
            n=input("Pretende usar o polinomio anterior? responda s para sim, n para nao:  ")
            if n=="s":
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                x=input("Introduza o numero de linhas a calcular:  ")
                tabela(p, x)
                
            elif n=="n":
                n2=input("Introduza o numero de termos do polinomio:")
                p=criarPolinomio()
                for i in range (int(n2)):
                    c=input("Introduza o coeficiente:")
                    e=input("Introduza o expoente:")
                    t=criarTermo(c, e)
                    inserirTermo(p, t)
                    verPolinomio(p)
                ordenarPol(p)
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                x=input("Introduza o numero de linhas a calcular:  ")
                tabela(p, x)
            
    elif op=="4":
        if p==[]:
            n2=input("Introduza o numero de termos do polinomio:")
            p=criarPolinomio()
            for i in range (int(n2)):
                c=input("Introduza o coeficiente:")
                e=input("Introduza o expoente:")
                t=criarTermo(c, e)
                inserirTermo(p, t)
                verPolinomio(p)
            ordenarPol(p)
            print("O polinómio que vai ser usado é: ")
            verPolinomio(p)
            print("O polinómio simplificado é: ")
            simplificarPol(p)
        
        else:
        
            print("O polinómio anterior é: ")
            verPolinomio(p)
            n=input("Pretende usar o polinomio anterior? responda s para sim, n para nao:  ")
            if n=="s":
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                print("O polinómio simplificado é: ")
                simplificarPol(p)
            elif n=="n":
                n2=input("Introduza o numero de termos do polinomio:")
                p=criarPolinomio()
                for i in range (int(n2)):
                    c=input("Introduza o coeficiente:")
                    e=input("Introduza o expoente:")
                    t=criarTermo(c, e)
                    inserirTermo(p, t)
                    verPolinomio(p)
                ordenarPol(p)
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                print("O polinómio simplificado é: ")
                simplificarPol(p)
                
            
    elif op=="5": 
        if p==[]:
            n2=input("Introduza o numero de termos do polinomio:")
            p=criarPolinomio()
            for i in range (int(n2)):
                c=input("Introduza o coeficiente:")
                e=input("Introduza o expoente:")
                t=criarTermo(c, e)
                inserirTermo(p, t)
                verPolinomio(p)
            ordenarPol(p)
            print("O polinómio que vai ser usado é: ")
            verPolinomio(p)
            p=derivarPol(p)
            simplificarPol(p)
          
            
        else:
            print("O polinómio anterior é: ")
            verPolinomio(p)
            n=input("Pretende usar o polinomio anterior? responda s para sim, n para nao:  ")
            if n=="s":
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                p=derivarPol(p)
                simplificarPol(p)
                
                
            elif n=="n":
                n2=input("Introduza o numero de termos do polinomio:")
                p=criarPolinomio()
                for i in range (int(n2)):
                    c=input("Introduza o coeficiente:")
                    e=input("Introduza o expoente:")
                    t=criarTermo(c, e)
                    inserirTermo(p, t)
                    verPolinomio(p)
                ordenarPol(p)
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                p=derivarPol(p)
                simplificarPol(p)
                
                
    elif op=="6":
        if p==[]:
            n2=input("Introduza o numero de termos do polinomio:")
            p=criarPolinomio()
            for i in range (int(n2)):
                c=input("Introduza o coeficiente:")
                e=input("Introduza o expoente:")
                t=criarTermo(c, e)
                inserirTermo(p, t)
                verPolinomio(p)
            ordenarPol(p)
            print("O polinómio que vai ser usado é: ")
            verPolinomio(p)
            num=input("quantos pontos quer que sejam calculados:  ")
            desenharGrafico(p, num)
          
            
        else:
            print("O polinómio anterior é: ")
            verPolinomio(p)
            n=input("Pretende usar o polinomio anterior? responda s para sim, n para nao:  ")
            if n=="s":
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                num=input("quantos pontos quer que sejam calculados:  ")
                desenharGrafico(p, num)
                
                
            elif n=="n":
                n2=input("Introduza o numero de termos do polinomio:")
                p=criarPolinomio()
                for i in range (int(n2)):
                    c=input("Introduza o coeficiente:")
                    e=input("Introduza o expoente:")
                    t=criarTermo(c, e)
                    inserirTermo(p, t)
                    verPolinomio(p)
                ordenarPol(p)
                print("O polinómio que vai ser usado é: ")
                verPolinomio(p)
                num=input("quantos pontos quer que sejam calculados:  ")
                desenharGrafico(p, num)
        
```
