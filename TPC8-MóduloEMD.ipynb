```python

def getReg(linha):
    linha1=linha.replace("\n", "")
    reg=linha1.split(",")
    return reg


def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    i=1
    for linha in f:
        linhanova=getReg(linha)
        linhanova=linhanova[1:]
        id="emd"+str(i)
        linhanova[0]=id
        bd.append(linhanova)
        i=i+1
    return bd


def chaveOrd(reg):
    return (reg[1])
    
    
def listarDataset(bd):
    print("id   | DataEMD |          Nome         | Resultado ")
    print("__________________________________________________")
    bd.sort(key=chaveOrd, reverse=True)
    for reg in bd:
        string=str(reg[0])+ "|" + str(reg[1])+ "|" + str(reg[2]) + " " + str(reg[3]) + "|" + str(reg[11])            
        print(string)             



def consultarDataset(bd, id):
    encontrado=False
    i=0
    while encontrado==False and i<(len(bd)):
        if bd[i][0]==id:
            #string=""
            #for elem in bd[i][0]:
            #    string=string+ str(elem)+ "|" 
            #print(string)
            encontrado=True
            
        else:
            encontrado=False
        i=i+1
    if encontrado==True:
        return bd[i]
    else:
        return encontrado
    
    
def modalidades(bd):
    lista=[]
    for elem in bd:
        if elem[7] not in lista:
            lista.append(str(elem[7])) 
    lista.sort()
    return lista


def distribPorModalidade(bd):
    distribuicaomod={}
    for elem in bd:
        if elem[7] in distribuicaomod.keys():
            distribuicaomod[elem[7]]=distribuicaomod[elem[7]]+1
        else:
            distribuicaomod[elem[7]]=1
    
    return distribuicaomod

def distribPorClube(bd):
    distribuicaoclub={}
    for elem in bd:
        if elem[8] in distribuicaoclub.keys():
            distribuicaoclub[elem[8]]=distribuicaoclub[elem[8]]+1
        else:
            distribuicaoclub[elem[8]]=1
    
    return distribuicaoclub

def distribPorAno(bd):
    distribuicaoano={}
    for elem in bd:
        data=elem[1]
        ano=data[:4]
        if ano in distribuicaoano.keys():
            distribuicaoano[ano]=distribuicaoano[ano]+1
        else:
            distribuicaoano[ano]=1
    
    return distribuicaoano


def distrib(bd, i):
    distrib={}
    for elem in bd:
        if elem[i] in distrib.keys():
            distrib[elem[i]]=distrib[elem[i]]+1
        else:
            distrib[elem[i]]=1
    return distrib

import matplotlib.pyplot as plt

def plotDistribuicaoPorModalidade(dist):
    x=[]
    for i in range (1, len (dist.keys())+1):
        x.append(i)
    y=[]
    labels=[]
    for curso in dist.keys():
        y.append (dist[curso])
        labels.append (curso)
    plt.bar(x,y, thick_label=labels, color=['red', 'green'])
    plt.xlabel('cursos')
    plt.ylabel('numero de alunos')
    plt.title('Distribuicao por cursos')
    plt.show()


def plotDistrib(dist):
    x=[]
    for i in range (1, len (dist.keys())+1):
        x.append(i)
    y=[]
    labels=[]
    for elem in dist.keys():
        y.append (dist[elem])
        labels.append (elem)
    plt.bar(x,y, thick_label=labels, color=['red', 'green'])
    plt.xlabel('')
    plt.ylabel('número')
    plt.title('Distribuicao ')
    plt.show()
    
    
 
```
